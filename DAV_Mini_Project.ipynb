# Interactive Graphs

# Cases with Country and timeline interactive country dropdown list

```python
from ipywidgets import widgets, interactive
```

```python
# Make a dropdown to select the Area, or "All"
area = widgets.Dropdown(
    options=['All'] + list(data['country'].unique()),
    value='All',
    description='Country:',
)
def plotit(area):
    df2 = data.copy()
    if area != 'All':
        df2 = df2[df2.country == area]
    df2 = df2[['dates','cases']]
    if len(df2) > 0:
        sns.lineplot(df2['dates'],df2['cases'])
        plt.show();
    else:
        print("No data to show for current selection")
```

```python
interactive(plotit, area=area)
```

# deaths with Country and timeline interactive country dropdown list

```python
# Make a dropdown to select the Area, or "All"
area = widgets.Dropdown(
    options=['All'] + list(data['country'].unique()),
    value='All',
    description='Country:',
)
def plotit(area):
    df2 = data.copy()
    if area != 'All':
        df2 = df2[df2.country == area]
    df2 = df2[['dates','deaths']]
    if len(df2) > 0:
        sns.lineplot(df2['dates'],df2['deaths'])
        plt.show();
    else:
        print("No data to show for current selection")
```

```python
interactive(plotit, area=area)
```

# Interactive cases and deaths Country wise barplots

```python
area = widgets.Dropdown(
    options=['All'] + list(data['country'].unique()),
    value='All',
    description='Country:'
)

curf = widgets.Dropdown(
    options=['All'] + list(data['curf'].unique()),
    value='All',
    description='curf:'
)

def plotit(area,curf):
    df2 = data.copy()
    if area != 'All':
        df2 = df2[df2.country == area]
    if curf!='All':
        df2=df2[df2.curf==curf]
        
    df2 = df2[['country','cases','deaths','curf']]
    if len(df2) > 0:
        df2.groupby('country')['cases','deaths'].sum().plot(kind='bar')
        plt.show()
    else:
        print("No data to show for current selection")
```

```python
interactive(plotit, area=area,curf=curf)
```

# No. of Cases and deaths monthly wise with country dropdown

```python
area = widgets.Dropdown(
    options=['All'] + list(data['country'].unique()),
    value='All',
    description='Country:'
)

month = widgets.Dropdown(
    options=['All'] + sorted(list(data['month'].unique())),
    value='All',
    description='Month:'
)

def plotit(area,month):
    df2 = data.copy()
    if area != 'All':
        df2 = df2[df2.country == area]
    if month!='All':
        df2=df2[df2.month==month]
        
    df2 = df2[['country','cases','deaths','curf','month']]
    if len(df2) > 0:
        df2.groupby('month')['cases','deaths'].sum().plot(kind='bar')
        plt.show()
    else:
        print("No data to show for current selection")
```

```python
interactive(plotit, area=area,month=month)
```
